```python
fig=plt.figure(figsize=(16,9))

fig,ax=plt.subplots(2,sharex=True)
data=infosys.iloc[:20]


#For Missing Data
data.asfreq('D').plot(ax=ax[0],marker='*')
ax[0].legend(['Missing Values'])


# After Filling 
data.asfreq('D',method='bfill').plot(ax=ax[1],style='-o')
data.asfreq('D',method='ffill').plot(ax=ax[1],style='--*')

ax[1].legend(['Backward Fill','Forward Fill']);
```

### Time Shifts

```python
fig,ax=plt.subplots(3,sharey=True)

infosys=infosys.asfreq('D',method='pad')

infosys.plot(ax=ax[0])
infosys.shift(900).plot(ax=ax[1])
infosys.tshift(900).plot(ax=ax[2])


local_max=pd.to_datetime('2010-12-05')
offset=pd.Timedelta(900,'D')



ax[0].legend(['input'],loc=2)
ax[0].get_xticklabels()[2].set(weight='heavy',color='red')
ax[0].axvline(local_max,alpha=0.5,color='red')


ax[1].legend(['shift'],loc=2)
ax[1].get_xticklabels()[2].set(weight='heavy',color='red')
ax[1].axvline(local_max + offset,alpha=0.5,color='green')


ax[2].legend(['Tshift'],loc=2)
ax[2].get_xticklabels()[2].set(weight='heavy',color='red')
ax[2].axvline(local_max + offset,alpha=0.5,color='y')
```

## ROI=Net Profit/Total Invesment*100

```python
fig=plt.figure(figsize=(16,8))
ROI=100* (infosys.shift(365)/infosys - 1 )
ROI.plot()
plt.ylabel('% Return on Invesment');
```

```python
fig=plt.figure(figsize=(16,8))
ROI=100* (infosys.shift(125)/infosys - 1 )
ROI.plot()
plt.ylabel('% Return on Invesment');
```

```python
plt.figure(figsize=(16,8))
rolling=infosys.rolling(365,center=False)

data=pd.DataFrame({'input':infosys,
                   '1 Year Rolling Mean':rolling.mean(),
                   '1 year Rolling Std':rolling.std()})

ax=data.plot(style=['-','--',':'])
ax.lines[0].set_alpha(0.5)
```

```python
data=pd.read_csv('https://raw.githubusercontent.com/reddyprasade/Data-Sets-For-Machine-Learnig-and-Data-Science/master/DataSets/Fremont_Bridge_Bicycle_Counter.csv',index_col='Date',parse_dates=True)
data.head()
```

```python
data.columns
```

```python
data.describe()
```

```python
data.info()
```

```python
data.plot()
plt.ylabel('Hourly Bicycle Count');
```

```python
weekly=data.resample('W').sum()
weekly.head()
```

```python
weekly.plot(style=['.-','--','-'])
plt.ylabel('Weekly Bicycle Count');
```

```python
Daily=data.resample('D').sum()
Daily.head()
```

```python
Daily.plot(style=['.-','--','-'])
plt.ylabel('Daily Bicycles Count');
```

```python
Daily.rolling(120,center=True).std().plot(style=['-','--',':'])  #  Std
plt.ylabel('Std hourly count');
```

```python
Daily.rolling(120,center=True).mean().plot(style=['-','--',':'])  # mean
plt.ylabel('mean hourly count');
```

```python
Daily.rolling(50,center=True,win_type='gaussian').sum(std=10).plot(style=['.-','--','-']);
```

```python
Daily.rolling(150,center=True,win_type='gaussian').sum(std=50).plot(style=['.-','--','-']);
```

```python
by_time=data.groupby(data.index.time).sum()
by_time.head()
```

```python
plt.figure(figsize=(16,9))
hourly_ticks=3*60*60*np.arange(8)
by_time.plot(xticks=hourly_ticks,style=[':','--','-']);
```

```python
by_weekday=data.groupby(data.index.dayofweek).sum()
by_weekday.head()
```

```python
by_weekday.index=['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
by_weekday.plot(style=[':','-','--']);
```

```python
weekend=np.where(data.index.weekday < 5 ,'weekday','weekend')
weekend[:50]
```

```python
by_time=data.groupby([weekend,data.index.time]).mean()
by_time.head()
```

```python
fig,ax=plt.subplots(1,2,figsize=(16,9))
by_time.loc['weekday'].plot(ax=ax[0],title='weekdays',xticks=hourly_ticks,style=[':','--','-'])
by_time.loc['weekend'].plot(ax=ax[1],title='weekends',xticks=hourly_ticks,style=[':','--','-']);
```
